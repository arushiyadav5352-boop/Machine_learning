```python
#One Hot Encoding 
```

```python
x_encoded=pd.get_dummies(df,columns=['model','transmission','fuelType'],drop_first=True)
```

```python
x_encoded
```

```python
x_encoded=x_encoded.astype(int)
```

```python
x_encoded
```

```python
#Standard Scaling 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df['year'] = df['year'].astype(int)
print(df['year'].dtype)
df.head()
```

```python

numerical_col = ['year', 'price', 'mileage', 'tax', 'mpg']

scaler = StandardScaler()

x_encoded[numerical_col] = scaler.fit_transform(x_encoded[numerical_col])
```

```python
x_encoded
```

```python
#Linear reggresion model 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= x_encoded.drop('price', axis=1)
```

```python
y= x_encoded['price']
```

```python
 x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred= model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
#Adjust r2
```

```python
n= x_test.shape[0]
p= x_test.shape[1]
adjust_r2= 1-((1-r2)*(n-1)/(n-p-1))
```
